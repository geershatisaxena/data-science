Catplot: Survival by class and gender

```python
sns.catplot(data=df, x='class', hue='sex', col='survived', kind='count', palette='viridis')
plt.suptitle('Survival by Class and Gender')
```

Histogram: Fare distribution

Boxenplot: Age by class

```python
sns.boxenplot(data=df, x='class', y='age', palette='cubehelix')
plt.title('Age Distribution by Class')
```

Jointplot: Age vs Fare

```python
import plotly.express as px
```

```python
df.head()
```

```python
px.scatter(df,x='age',y='fare',color='survived',title='Age vs Fare by Survival',template='plotly_dark')
```

```python
px.sunburst(df,path=['pclass','sex','survived'],values='age',
            title='Survived by class and gender',width=500,height=500,
            template='plotly_dark',color='age',color_continuous_scale='plasma')
```

```python
px.histogram(df,x='age',nbins=30,color='survived',title='Age Distribution by Survival Status',template='plotly_dark')
```

```python
px.treemap(df,path=['pclass','survived'],values='age',title='Survival by class',
           width=800,height=500,template='plotly_dark',color='age',color_continuous_scale='viridis')
```

```python
survived_counts = df['survived'].value_counts()
px.pie(names=survived_counts.index, values=survived_counts.values,
       title='Survival Rate',template='plotly_dark',
       width=500,height=500)
```

```python
fig1 = px.bar(df, x='sex', color='survived', title='Count of Survivors by Gender', 
               labels={'survived': 'Survived'}, color_discrete_sequence=['red', 'green'],template='plotly_dark')
fig1.show()
```

```python
fig2 = px.bar(df, x='class', color='survived', title='Count of Survivors by Class', 
               labels={'survived': 'Survived'}, color_discrete_sequence=['red', 'green'])
fig2.show()
```

```python
fig3 = px.bar(df, x='embark_town', title='Count of Passengers by Embarkation Town')
fig3.show()
```

```python
fig4 = px.bar(df.groupby('class')['age'].mean().reset_index(), x='class', y='age', 
               title='Average Age by Class')
fig4.show()
```

```python
fig5 = px.bar(df.groupby('class')['fare'].mean().reset_index(), x='class', y='fare', 
               title='Average Fare by Class')
fig5.show()
```

```python
# Scatter plot: Age vs Fare with size based on number of parents/children aboard
fig4 = px.scatter(df, x='age', y='fare', size='parch', color='survived', 
                  title='Age vs Fare with Parents/Children Count', 
                  labels={'parch': 'Parents/Children Aboard'})
fig4.show()
```

```python
# Scatter plot: Age vs Fare with size based on number of siblings/spouses aboard
fig3 = px.scatter(df, x='age', y='fare', size='sibsp', color='survived', 
                  title='Age vs Fare with Siblings/Spouses Count', 
                  labels={'sibsp': 'Siblings/Spouses Aboard'})
fig3.show()
```

```python
# Scatter plot: Age vs Fare with a custom marker size
fig15 = px.scatter(df, x='age', y='fare', size_max=20, 
                   title='Age vs Fare with Custom Marker Size')
fig15.show()
```
