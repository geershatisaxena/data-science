```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
sns.countplot(df,y='survived',color='r')
plt.title('Count of Survived vs Not Survived')
```

```python
sns.countplot(df,x='survived',hue='sex',palette='ocean')
plt.title('Count of Survived vs Not Survived by gender')
```

```python
sns.barplot(df,x = 'class',y = 'survived',palette='Dark2')
plt.title('Survival Rate by Class')
```

```python
sns.histplot(df,x='age',bins = 30,kde=True,color= 'purple')
plt.title('Age Distribution of Passengers')
plt.show()
```


 1. Countplot: Survival count

```python
sns.countplot(data=df, x='survived', color='blue')
plt.title('Survival Count')
```

Countplot: Survival by gender

```python
sns.countplot(data=df, x='survived', hue='sex', palette='bwr')
plt.title('Survival by Gender')
```

Barplot: Survival rate by class

```python
sns.barplot(data=df, x='class', y='survived', palette='plasma')
plt.title('Survival Rate by Class')
```

Histogram: Age distribution

```python
sns.histplot(data=df, x='age', bins=30, kde=True, color='peru')
plt.title('Age Distribution')
```

Boxplot: Age by class

```python
sns.boxplot(data=df, x='class', y='age', palette='Set2')
plt.title('Age Distribution by Class')
```

Violinplot: Age by gender

```python
sns.violinplot(data=df, x='sex', y='age', hue='survived', split=True, palette='muted')
plt.title('Age Distribution by Gender and Survival')
```

Pairplot: Numerical features

```python
sns.pairplot(df, hue='survived', palette='husl')
plt.title('Pairplot of Numerical Features')
```

Swarmplot: Fare by class

```python
sns.swarmplot(data=df, x='class', y='fare', hue='sex', palette='Set1')
plt.title('Fare Distribution by Class and Gender')
plt.show()
```

Stripplot: Age by survival

```python
sns.stripplot(data=df, x='survived', y='age', jitter=True, hue='sex', palette='ocean')
plt.title('Age by Survival')
```

KDEplot: Age distribution by survival

```python
sns.kdeplot(data=df, x='age', hue='survived', fill=True, palette='Reds')
plt.title('Age Distribution by Survival')
```

Barplot: Average fare by class

```python
sns.barplot(data=df, x='class', y='fare', palette='inferno')
plt.title('Average Fare by Class')
```

Countplot: Embarkation points

```python
sns.countplot(data=df, x='embark_town', palette='magma')
plt.title('Passenger Count by Embarkation Point')
```

Boxplot: Fare by embarkation point

```python
sns.boxplot(data=df, x='embark_town', y='fare', palette='cool')
plt.title('Fare Distribution by Embarkation Point')
```

Violinplot: Fare by survival

```python
sns.violinplot(data=df, x='survived', y='fare', palette='magma')
plt.title('Fare Distribution by Survival')
```

Catplot: Survival by class and gender

```python
sns.catplot(data=df, x='class', hue='sex', col='survived', kind='count', palette='viridis')
plt.suptitle('Survival by Class and Gender')
```

Histogram: Fare distribution

```python
sns.histplot(data=df, x='fare', bins=40, kde=True, color='crimson')
plt.title('Fare Distribution')
```

KDEplot: Fare by survival

```python
sns.kdeplot(data=df, x='fare', hue='survived', fill=True, palette='Reds')
plt.title('Fare Distribution by Survival')
```

Boxenplot: Age by class

```python
sns.boxenplot(data=df, x='class', y='age', palette='cubehelix')
plt.title('Age Distribution by Class')
```

Jointplot: Age vs Fare


```python
sns.jointplot(data=df, x='age', y='fare', hue='survived', kind='scatter', palette='magma')
plt.suptitle('Age vs Fare by Survival')
```
