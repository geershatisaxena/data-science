```python
fig,ax = plt.subplots(figsize=(5,3))
fig.set_facecolor('lightblue')
fig.set_facecolor('cyan')
#plotting the data
x= [10,25,85,96,85,45,63]
y= [10,25,85,96,5,45,63]
ax.plot(x,y,marker ='s',
        markersize = 10,
        markerfacecolor = 'r', 
        markeredgecolor = 'k',
        markeredgewidth = 2,
        linestyle = '-.',
        color = 'g',
       
        linewidth = 2)
#setting the title and labels
ax.set_title('Sample plot', fontsize=16,color = 'r')
ax.set_xlabel('X-axis', fontsize=14,color = 'b')
ax.set_ylabel('Y-axis', fontsize=14,color = 'b')

ax.axhline(y=50,color='k',linewidth=2.5,linestyle=':')
ax.axvline(x=50,color='k',linewidth=2.5,linestyle=':')

ax.text(15,65,'Sample Text',fontsize = 15,color = 'r')
```

```python

plt.plot([1,2,3,4,5],[21,12,32,6,45],marker = 's',label = 'y=x^2',color = 'gold')
plt.plot([2,5,1,4,6],[41,53,23,51,12],marker = 'd',label = 'y=x^3',color = 'g')
plt.xticks([1,2,3,4,5],['A','B','C','D','E'])
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
